```python
import random
import math
import time

def create_list():
    return [5, 1, 2, 4, 3]

def random_list(n=1000, sample_max=10_000):
    return random.sample(range(0, sample_max), n)
```

```python
def insert_sort(L):
    n = len(L)
    for i in range(1, n):
        cle = L[i]
        j = i-1
        while j >= 0 and L[j] > cle:
            L[j+1] = L[j]
            j -= 1
        L[j+1] = cle
    return L
```

```python
%timeit insert_sort(random_list())
```

```python
def fusion_sort(L):
    def fusion(L1, L2):
        L = []
        if L1 == []:
            return L2
        if L2 == []:
            return L1
        if L1[0] <= L2[0]:
            return [L1[0]] + fusion(L1[1:], L2)
        else:
            return [L2[0]] + fusion(L1, L2[1:])
    n = len(L)
    if n <= 1:
        return L
    m = math.floor(n/2)
    L1 = L[0: m]
    L2 = L[m:]
    return fusion(fusion_sort(L1), fusion_sort(L2))
```

```python
%timeit fusion_sort(random_list())
```

```python
debut = time.time()
fusion_sort(random_list())
fin = time.time()
print(fin-debut)
```

```python
def bubble_sort(L):
    n = len(L)
    for i in range(n-1, 0, -1):
        count = 0
        for j in range(0, i):
            if L[j] > L[j+1]:
                L[j], L[j+1] = L[j+1], L[j]
                count += 1
        if count == 0:
            break
    return L
```

```python
%timeit bubble_sort(Lp)
```

```python
def selection_sort(L):
    for i in range(0, len(L)):
        indicemin = i
        for j in range(i, len(L)):
            if L[j] < L[indicemin]:
                indicemin = j
        L[i], L[indicemin] = L[indicemin], L[i]
    return L
```

```python
%timeit selection_sort(random_list())
```

```python
def bucket_sort(L, max_val, dist):
    urnes = [[] for i in range(0, max_val//dist)]
    for e in L:
        for i in range(0, max_val//dist):
            if dist*i < e <= dist*(i+1)+1:
                urnes[i].append(e)
                break
           
    urnes = [insert_sort(urne) for urne in urnes]
    triee = urnes[0]
    for urne in urnes[1:]:
        if urne == []:
            continue
        if urne[0] < triee[0] :
            triee = urne + triee
        else:
            triee += urne
    return triee
```

```python
%timeit bucket_sort(random_list(), 10_000, 100)
```
